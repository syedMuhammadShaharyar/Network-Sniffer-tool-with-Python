```python
import tkinter as tk
from tkinter import scrolledtext
import threading
from scapy.all import sniff, IP
from PIL import Image, ImageTk  # You might need to run: pip install pillow

class SnifferGUI:
    def __init__(self, root):
        self.root = root
        self.root.title("Shahary Network Analyzer")
        self.root.geometry("800x600")
        self.bg_path = r"C:\Users\Shahary\Downloads\360_F_162765779_AQ9TlSDolclHQvgKceJZbbIq5wzuv63w.jpg"
        try:
            self.bg_image = Image.open(self.bg_path)
            self.bg_image = self.bg_image.resize((800, 600), Image.Resampling.LANCZOS)
            self.bg_photo = ImageTk.PhotoImage(self.bg_image)
            
            self.bg_label = tk.Label(root, image=self.bg_photo)
            self.bg_label.place(x=0, y=0, relwidth=1, relheight=1)
        except Exception as e:
            print(f"Error loading image: {e}")
            self.root.configure(bg="#1e1e1e")
        self.container = tk.Frame(root, bg="#000000", bd=2, relief="ridge")
        self.container.place(relx=0.5, rely=0.5, anchor="center", width=700, height=500)

        self.label = tk.Label(self.container, text="Shahary PACKET ANALYZER", 
                              font=("Courier", 16, "bold"), fg="#00FF00", bg="#000000")
        self.label.pack(pady=15)

        self.start_button = tk.Button(self.container, text="INITIALIZE SNIFFING", 
                                      command=self.start_thread, font=("Courier", 10, "bold"),
                                      bg="#004400", fg="white", activebackground="#008800", padx=20)
        self.start_button.pack(pady=10)
        self.display_area = scrolledtext.ScrolledText(self.container, width=80, height=18, 
                                                      font=("Courier", 10), bg="#0a0a0a", 
                                                      fg="#00FF00", insertbackground="white")
        self.display_area.pack(pady=10, padx=20)

        self.is_sniffing = False

    def packet_callback(self, packet):
        if packet.haslayer(IP):
            src = packet[IP].src
            dst = packet[IP].dst
            proto = "TCP" if packet.haslayer("TCP") else "UDP" if packet.haslayer("UDP") else "IP"
            info = f"[DETECTED] {src} -> {dst} | PROTO: {proto}\n"
            self.root.after(0, self.update_display, info)

    def update_display(self, text):
        self.display_area.insert(tk.END, text)
        self.display_area.yview(tk.END)

    def start_sniffing(self):
        self.is_sniffing = True
        sniff(prn=self.packet_callback, filter="ip", store=0)

    def start_thread(self):
        if not self.is_sniffing:
            self.start_button.config(state="disabled", text="[ MONITORING ACTIVE ]", bg="#880000")
            sniffer_thread = threading.Thread(target=self.start_sniffing)
            sniffer_thread.daemon = True
            sniffer_thread.start()

if __name__ == "__main__":
    root = tk.Tk()
    gui = SnifferGUI(root)
    root.mainloop()
```
